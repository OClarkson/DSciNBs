### Dropna and astype

```python
nba.dropna()
```

```python
nba.dropna(how='all', inplace=True) #removes rows where all values are nan only
```

```python
nba.dropna(axis=1) #removes columns where there were any nans
```

```python
nba.dropna(subset=['Salary']) #rows removed where Nans present in Salary column
```

```python
#fill nulls with a value
nba.fillna(value=0)
```

```python
nba["Salary"].fillna(0, inplace=True)
```

```python
nba
```

```python
nba["College"].fillna("None Listed", inplace=True)
```

```python
nba
```

```python
nba = pd.read_csv('./data/pandas/nba.csv').dropna(how ='all')
nba["Salary"].fillna(0, inplace=True)
nba["College"].fillna("None", inplace=True)
```

```python
nba.head(6)
```

```python
nba.info()
```

```python
nba['Salary'] = nba['Salary'].astype('int')
nba['Age'] = nba['Age'].astype('int')
nba['Number'] = nba['Number'].astype('int')
```

```python
nba.info()
```

```python
nba['Position'].nunique()
```

```python
nba['Position'] = nba['Position'].astype('category')
```

```python
nba['Position']  #looks the same but it stored differently in memory 
```

```python
nba['Team'] = nba['Team'].astype('category') 
```

```python
nba.info()
```
