### Broadcasting operations

```python
nba = pd.read_csv('./data/pandas/nba.csv')
```

```python
nba['Age'].add(500)
```

```python
nba['Age'] + 5
```

```python
nba['Salary'].sub(5e6)
```

```python
nba['Weight in Kg'] = nba['Weight'].mul(0.453592)
```

```python
nba
```

```python
nba['Number'] = nba['Number'].astype('Int64')
```

```python
nba
```

```python
nba['Team'].value_counts()
```

```python
nba['Team'].value_counts().head(1)
```

```python
nba['Salary'].sort_values(ascending=False)
```

```python
nba.dropna()
```

```python
nba.dropna(how='all', inplace=True) #removes rows where all values are nan only
```

```python
nba.dropna(axis=1) #removes columns where there were any nans
```

```python
nba.dropna(subset=['Salary']) #rows removed where Nans present in Salary column
```

```python
#fill nulls with a value
nba.fillna(value=0)
```
